### Data Cleaning

```python
# Dropping the pincode feature
data.drop('Postal Code', axis=1, inplace=True)
data.head()
```

### Exploratory Data Analysis

```python
# Pair plot
sns.pairplot(data,hue='Profit')
```

```python
#Correlation Plot
sns.heatmap(data.corr(), annot=True );
```

**Observation:-** <br>
- **Strong Correlation** is present between Independent feature Discount and Quantity.  <br>
- **Negatively Correlation** is present between Discount and Profit. <br>
- **Positive Correlation** is present between Sales and Profit. 

```python
sns.regplot(x="Sales", y="Profit", data=data)
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('Sales vs Profit')
```

**Observation**
- There is positive correlation between Sales and Profit, indicating that higher sales lead to the higher profits. <br>
- Apart from that there are many data point where the sales are high but the profits are either low or negative.  <br>
- This data visualization suggests that the company have to identify and address the factors that are reducing the profit margin.

```python
#Data Visualization between Sub-Category Vs Category
plt.figure(figsize=(20,10))
plt.bar('Sub-Category', 'Category', data=data)
plt.title('Sub-Category VS Category')
plt.xlabel('Sub-Category')
plt.ylabel('Category')
```

**Observation**
- We have 4 high category product in Phones, Accessories, Machines and Copies.

```python
data['Segment'].value_counts()
```

```python
data_segment= data.groupby(['Segment'])[['Sales', 'Discount', 'Profit']].mean()
data_segment
```

```python
#1. Sales 2. Discount 3. Profit
data_segment.plot.pie(subplots=True, 
                    autopct='%1.1f%%',
                    figsize=(18, 20),
                    startangle=90,     # start angle 90° (Africa)
                    shadow=True,
                    labels = data_segment.index)
plt.title('Segment wise analysis of Sale, Discount, Profit')
plt.show()
```
